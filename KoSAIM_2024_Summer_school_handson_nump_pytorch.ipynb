# Autograd

> > Central to all neural network in PyTorch is the `autograd` package

> > The `autograd` package provides automatic differentiation for all operations on Tensors.

> >`torch.Tensor` is the central class of package. If you set its attribute `.requires_grad` as True. it starts to track all operations on it. When you finish your computation you can call `.backward()` and have all the gradients computed automatically. The gradeint for this tensor will be accumulated in `.grad` attribute.

> > To stop a tensor from tracking history, you can call `.detach()` to detach it from the computation history.

> >  <img src='https://drive.google.com/uc?export=download&id=1D4Eh4UKYO692vYLezQ9y_UrGJ6kTuZQE' width="" height ="" /><br>

> > Example)
> > > $z = 2x^2 + 3$

# Linear Regression (nn.Module)

> > - Define Data (train dataset, test dataset)
> > > - Height(x) = [160, 170, 185, 190]
> > > - Weight(y) = [60, 70, 85, ?]

> > - Define Model
> > > - y = Wx + b


> > - Example) Height -> Weight

> > > - Mean Square Error

> > > image.png


> > > image.png

> > - Optimizer
> > > Find optimal W,b that minimize loss function

> > image.png

image.png

```python
# Define dataset
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])

# y = wx + b --> y = 2 * x + 0

# model initialize
W = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

# optimzier
optimizer = torch.optim.SGD([W, b], lr=0.001)

total_epochs = 1000
for epoch in range(total_epochs + 1):

    # predict model
    logit = x_train * W + b

    # calcuate loss
    # Quiz, MSE loss
    loss = torch.mean((logit - y_train)**2)

    # gradient update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format( epoch, total_epochs, W.item(), b.item(), loss.item()))
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
torch.manual_seed(123)
```

```python
# define x_train, y_train
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])

y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])

print('x_train:', x_train.shape)
print('y_train:', y_train.shape)
```

```python
# Quiz, Linear layer from scratch
```

```python
# input_dim = 3, output_dim = 1
model = nn.Linear(3,1)

# y = wx + b
print('model:', model, '\n')
model_params = list(model.parameters())
print('weight: \n', model_params[0], '\n')
print('bias: \n', model_params[1], '\n')
```

```python
# define optimizer,
optimizer = torch.optim.SGD(model.parameters(), lr=1e-5, )
```

```python
# training ...
total_epochs = 10
for epoch in range(total_epochs+1):

    # predict model
    prediction = model(x_train)


    # loss
    # loss = torch.mean((logit - y_train)**2)
    loss = F.mse_loss(prediction, y_train)  # Mean Square Error

    # set grad. to 0
    optimizer.zero_grad()
    # calculate gradient
    loss.backward()
    # update weight, bias
    optimizer.step()

    if epoch % 2 == 0:
      print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, total_epochs, loss.item()))
```

```python
pred_y = model(x_train[0])
print('model prediction :', pred_y)
```

```python
# updated model
updated_model_params = list(model.parameters())
print('weight: \n', updated_model_params[0], '\n')
print('bias: \n', updated_model_params[1], '\n')
```

```python
# Quiz, why optimizer.zero_grad()?

w = torch.tensor(2.0, requires_grad=True)

totla_epochs = 20
for i in range(totla_epochs):
    z = 2*w
    optimizer.zero_grad()
    z.backward()
    optimizer.step()

    print('w.grad:', w.grad)
```

```python
# define x_train, y_train
x_data = torch.FloatTensor([[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]])
y_data = torch.FloatTensor([[0], [0], [0], [1], [1], [1]])
```

```python
# define model
model = nn.Sequential(
    nn.Linear(2,1),
    nn.Sigmoid()
)
model
```

```python
# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)
```

```python
# logits
logits = model(x_data)
logits
```

```python
# training...

total_epochs = 1000
for epoch in range(total_epochs + 1):

    # predict model
    logits = model(x_data)

    # calcuate loss
    loss = F.binary_cross_entropy(logits, y_data)

    # update gradient
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        prediction = logits >= torch.FloatTensor([0.5]) # Prediction > 0.5 --> True
        correct_prediction = prediction.float() == y_data # Prediction == Ground Truth --> True
        accuracy = correct_prediction.sum().item() / len(correct_prediction) # cal. accuracy
        print('Epoch {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(epoch, total_epochs, loss.item(), accuracy * 100,))
```

```python
predicted = model(x_data) > torch.FloatTensor([0.5])
predicted = [[1] if i == True else [0] for i in predicted ]
print('predicted:', predicted)

print('true labe: \n', y_data)
```

```python
# define x_train, y_train
x_train = [[1, 2, 1, 1],
           [2, 1, 3, 2],
           [3, 1, 3, 4],
           [4, 1, 5, 5],
           [1, 7, 5, 5],
           [1, 2, 5, 6],
           [1, 6, 6, 6],
           [1, 7, 7, 7]]
y_train = [2, 2, 2, 1, 1, 1, 0, 0]  # 3 classes, 0, 1, 2
x_train = torch.FloatTensor(x_train)
y_train = torch.LongTensor(y_train)

print('x_train: \n', x_train, '\n')
print('x_train.shape:', x_train.shape)  # 8 samples with 4 features
print('\n')
print('y_train: \n', y_train, '\n')
print('y_train.shape:', y_train.shape)
```

> > # Quiz, Why One-hot encoding?

> > Example) 3 classes
> > > - {banana, apple, tomato}
> > > - loss

> > > > image.png

```python
# banana, apple, tomato -> 1, 2, 3

# [[1, 0,0],
#  [0,1, 0],
#  [0,0,1]]]

# GT: banana (1), predcit:apple (2) -> (1- 2) ^ 2 -> 1
# GT: banana (1), predcit:tomato (3). -> (1 -3 ) ^ 2 -> 4
```

```python
y_one_hot = torch.zeros(8, 3)
print(y_train)
print(y_train.unsqueeze(1))
y_one_hot.scatter_(1, y_train.unsqueeze(1), 1)
print('y_one_hot: \n', y_one_hot, '\n')
print('y_one_hot.shape:', y_one_hot.shape)
```

```python
# define model
class SoftmaxClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(4,3)
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, x):
        return self.softmax(self.linear(x))

# Quiz, Multi-layer perceptron
```

```python
# init model, define optimizer
model = SoftmaxClassifier()
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

```python
print(x_train)
print(x_train.shape)
print(y_train)
print(y_train.shape)
```

```python
# training ...
total_epochs = 1000
for epoch in range(total_epochs + 1):
    correct = 0
    # predict model
    prediction = model(x_train)

    # cal. loss
    #cross_entropy_loss=softmax+nll
    loss = F.cross_entropy(prediction, y_train) # softmax + nll

    # update gradeitn
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        _, predicted = torch.max(prediction, 1)
        correct += predicted.eq(y_train).sum()
        print('Epoch {:4d}/{} Cost: {:.6f} Acc, : {:.2f} %'.format(epoch, total_epochs, loss.item(), correct / len(y_train) * 100))
```
